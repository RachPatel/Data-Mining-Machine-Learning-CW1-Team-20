In the code below , Whta we do will adress reducing the dataSet to a smaller size using of of the sklearn ensembles ExtraTreesClassifer.We have the option of selecting n features which contribute the most towards the target variable.
This will select the features based on the multiclassfication problem at hand.The Otput is a bar plot wwith features on the y axia and the corresponding  value of contribution towards the target class.From the diagram we can see according to extraTreesClassifer the most contributing is feature 1030.

```python
import pandas as pd
import numpy as np

X = features_plus_classes.iloc[:,0:2303]  #independent columns
y = features_plus_classes.iloc[:,-1]    #target column 

from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) 

#use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization

feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

#  Running Naive Bayes on the Tranformed DataSet

# implementation of Naive bayes 1.1: 10 Features

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
model = GaussianNB()
#define the feature vector

features_plus_classes_Tranformed_features=features_plus_classes_Tranformed10.loc[:, features_plus_classes_Tranformed10.columns != 'Class']
features_plus_classes_Tranformed_labels=y = features_plus_classes_Tranformed10.iloc[:,-1]  
X_train, X_test, y_train, y_test = train_test_split( features_plus_classes_Tranformed_features, features_plus_classes_Tranformed_labels, test_size=0.2, random_state=0)
Model1=model.fit(X_train,y_train)

#Predictions on test data
y_pred=Model1.predict(X_test)
#print("Number of mislabeled points out of a total %d points : %d"(X_test.shape[0], (y_test != y_pred).sum()))
print("total:",X_test.shape[0],"Misclassified:",(y_test != y_pred).sum())
accuracy_score(y_test,y_pred)
```
